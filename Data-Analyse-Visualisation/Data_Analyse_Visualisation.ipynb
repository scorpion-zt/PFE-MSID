```python
#pip install pandas matplotlib

import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Chemin vers le Fichier CSV

path_to_csv_file = '/Users/mac/Downloads/PFE-MSID/Data-Input/data_News.csv'

#Lecture du Fichier CSV

df = pd.read_csv(path_to_csv_file,delimiter=';')
```

```python
#Affichage des 10 premières lignes du DataFrame pour visualiser un échantillon des données

print(df.head(10))
```

```python
#Affichage des informations sur le DataFrame, y compris le type de données de chaque colonne et les valeurs manquantes

print(df.info())
```

```python
#Affichage des statistiques descriptives sur les données, telles que la moyenne, l'écart type, le minimum et le maximum

print(df.describe())
```

```python
#Diagramme à barres pour visualiser la distribution des origines des nouvelles

plt.figure(figsize=(8, 6))
df['Origin'].head(200).value_counts().plot(kind='bar')
plt.title('Distribution des Origines des Nouvelles')
plt.xlabel('Origine')
plt.ylabel('Nombre de Nouvelles')
plt.show()
```

```python
#Diagramme circulaire pour visualiser la distribution des origines des nouvelles

plt.figure(figsize=(8, 6))
df['Origin'].head(200).value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Diagramme Circulaire des Origines des Nouvelles')
plt.ylabel('')
plt.show()
```
